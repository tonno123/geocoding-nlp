Ik kan de resultaten nog niet op een kaart plotten, het makkelijkst is om een rij coördinaten te kopiëren naar http://www.hamstermap.com/quickmap.php om ze te bekijken. 

Als output geeft het programma twee categorieën, namelijk 'road coordinates' en 'bounding box'. Road coordinates is een lijst met coördinaten per gevonden weg. Hierin staan ook coördinaten van kruispunten. Bounding box zijn twee coördinaten van een plaats, dorp of stad die is voorgekomen in het artikel.

Het programma zoekt iedere locatie op via Nominatim of Overpass. Nominatim wordt gebruikt voor plaatsen, dorpen, steden, etc. Overpass wordt gebruikt voor wegen. Als er online een resultaat wordt gevonden, wordt dat vergeleken met andere locaties in de zin, als die er zijn. Zo worden locaties gecombineerd en de resultaat coördinatenlijst steeds korter. Dit wordt per zin gedaan, en aan het eind worden alle gevonden locaties in het artikel met elkaar vergeleken. Hierdoor is het vaak zo dat de 'bounding box'-categorie wegvalt, omdat deze wordt gecombineerd met een straat of snelweg.
